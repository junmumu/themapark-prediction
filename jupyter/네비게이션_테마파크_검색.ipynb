```python
import requests
import bs4
import datetime as dt
import pandas as pd
```

```python
# url = 'https://datalab.visitkorea.or.kr/datalab/portal/bda/getTourVisitCnt.do#'
url = 'https://datalab.visitkorea.or.kr/visualize/getTempleteData.do'
params = {
    'txtSGG_CD': '1',
    'txtSIDO_ARR': '',
    'SGG_CD': '98',
    'TMAP_CATE_MCLS_CD':'문화관광',
    'SIDO_ARR': '',
    'BASE_YM1': '20191006',
    'BASE_YM2': '20191006',
    'srchAreaDate': '5',
    'qid': 'BDT_03_04_002'
}
# 월별 : srchAreaDate : 1
```

```python
def cal_std_month(before_month):
    x = dt.datetime.now() - dt.timedelta(30*before_month) #현재날짜-지정날짜
    year = x.year #연도
    month = x.month if x.month >= 10 else '0'+ str(x.month) #월
    return str(year) + str(month)
```

```python
def cal_std_day(before_day):
    x = dt.datetime.now() - dt.timedelta(before_day) #현재날짜-지정날짜
    year = x.year #연도
    month = x.month if x.month >= 10 else '0'+ str(x.month)
    day = x.day if x.day >= 10 else '0'+ str(x.day)
    return str(year) + str(month) + str(day)
```

```python
cal_std_day(1748)
cal_std_day(107)
cal_std_day(1105)
cal_std_day(107)
```

```python
response = requests.post(url, data=params)
res = response.json()
res
```

```python
data = {'에버랜드':[], '서울대공원': [], '어린이대공원': [], '롯데월드잠실점': []}
data2 = []
index = []
```

```python
for i in range(1105, 1104, -1):  # 1748, 106, -1 / 1105, 1104 -1
#     t = []
#     t.append(day)
    day = cal_std_day(i)
    index.append(day)
    params['BASE_YM1'] = day
    params['BASE_YM2'] = day
    response = requests.post(url, data=params)
    res = response.json()
    for tmp in res['list']:
        if tmp['ITS_BRO_NM'] == '에버랜드':
            data['에버랜드'].append(tmp['SRCH_CNT'])
#             t.append(tmp['SRCH_CNT'])
#             data2.append(t)
        elif tmp['ITS_BRO_NM'] == '서울대공원':
            data['서울대공원'].append(tmp['SRCH_CNT'])

        elif tmp['ITS_BRO_NM'] == '어린이대공원':
            data['어린이대공원'].append(tmp['SRCH_CNT'])

        elif tmp['ITS_BRO_NM'] == '롯데월드잠실점':
            data['롯데월드잠실점'].append(tmp['SRCH_CNT'])
```

```python
print(len(data['에버랜드']))
print(len(data['서울대공원']))
print(len(data['어린이대공원']))
print(len(data['롯데월드잠실점']))
print(len(data2))
print(len(index))
```

```python
df = pd.DataFrame(data, index=index)
df
```
