### 롯데월드

```python
def cal_std_day(before_day):
    x = dt.datetime.now() - dt.timedelta(before_day) #현재날짜-지정날짜
    year = x.year #연도
    month = x.month if x.month >= 10 else '0'+ str(x.month)
    day = x.day if x.day >= 10 else '0'+ str(x.day)
    return str(year) + str(month) + str(day)
```

```python
df_lotfest = pd.read_csv('./롯데월드축제정보.csv', encoding='cp949')
```

```python
df_lotfest
```

```python
lotfest = list(df_lotfest.itertuples(index=False, name=None))
```

```python
lotfest
```

```python
lotfest_dict = {'날짜': [], '행사ox': [], '행사명': []}
```

```python
for i in range(1748, 106, -1):  # 1748, 106, -1 
    day = cal_std_day(i)
    lotfest_dict['날짜'].append(day)
    for k in range(len(lotfest)):
        startdate = str(lotfest[k][1])
        enddate = str(lotfest[k][2])
        date1 = dt.datetime(int(startdate[:4]), int(startdate[4:6]), int(startdate[6:8]))
        date2 = dt.datetime(int(enddate[:4]), int(enddate[4:6]), int(enddate[6:8]))
        d = datetime(int(day[:4]), int(day[4:6]), int(day[6:8]))
        if date1 <= d <= date2:
            lotfest_dict['행사ox'].append('o')
            lotfest_dict['행사명'].append(lotfest[k][0])
            break
    else:
        lotfest_dict['행사ox'].append('x')
        lotfest_dict['행사명'].append('')
```

```python
df_lotfest_fin = pd.DataFrame(lotfest_dict)
```

```python
df_lotfest_fin
```

```python
df_lotfest_fin.to_csv('롯데월드_축제정보.csv', index=False, encoding='cp949')
```
